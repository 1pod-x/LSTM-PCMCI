```python
# Open connection for Coinbase WebSocket
def on_open_coinbase(ws, trading_pair):
    subscribe_message = {
    "type": "subscribe",
    "channels": ["level2"],
    "product_ids": [trading_pair]
    }
    ws.send(json.dumps(subscribe_message))
    print(f"Subscribed to Coinbase trade for {trading_pair}")
```

```python
# Open connection for Bitfinex WebSocket
def on_open_bitfinex(ws, trading_pair):
    subscribe_message = {
        "event": "subscribe",
        "channel": "book",
        "symbol": trading_pair
    }
    ws.send(json.dumps(subscribe_message))
```

```python
# Start WebSocket connection for Coinbase
def start_coinbase_ws(trading_pair):
    ws = websocket.WebSocketApp(coinbase_urls, 
                                on_message=on_message_coinbase, 
                                on_open=lambda ws: on_open_coinbase(ws, trading_pair))
    ws.run_forever(sslopt={"cert_reqs": ssl.CERT_NONE})
```

```python
# Start WebSocket connection for Binance
def start_binance_ws(trading_pair):
    binance_url = f"wss://stream.binance.com:9443/ws/{trading_pair}@trade"
    ws = websocket.WebSocketApp(binance_url, 
                                on_message=on_message_binance, 
                                on_open=lambda ws: on_open_binance(ws, trading_pair))
    ws.run_forever(sslopt={"cert_reqs": ssl.CERT_NONE})
```
